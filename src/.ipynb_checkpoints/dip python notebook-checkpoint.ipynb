```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])

#Below can also be done using 'mpimg.imread'. Gives a significantly different output.
#Change img path below
img = mpimg.imread('../data/srd/board7.jpeg')
gray = rgb2gray(img)

plt.imshow(gray, cmap = plt.get_cmap('gray'))

#plt.savefig('lena_greyscale_mpimg.png')
#plt.show()
```

#### Histogram Equalization

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = cv.imread('../data/board5.jpg',0)

# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)

cv.imwrite('clahe_2.jpg',cl1)
```
